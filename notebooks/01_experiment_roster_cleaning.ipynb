```python
## Loading the experiment roster
df_roster = pd.read_csv("df_final_experiment_clients.csv")
```

```python
### Check group distribution
#Counting clients per Variation (Test vs. Control) to confirm the groups are balanced enough for a fair comparison.
df_roster["Variation"].value_counts()
```

```python
#### Verify the primary key
#client_id should uniquely identify each client. We confirm this with `is_unique` and double-check by counting duplicates.
df_roster['client_id'].is_unique
df_roster["client_id"].duplicated().value_counts()
```

```python
### Check for missing values
#Counting nulls in `Variation` specifically and the overall missing-value percentage per column.
df_roster['Variation'].isna().sum()
df_roster.isna().mean() * 100
```

## Cleaning decisions

After exploring `df_roster`, no cleaning was applied at this stage:
